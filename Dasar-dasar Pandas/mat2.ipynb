```python
import pandas as pd
import numpy as np
```

```python
n_baris=10
n_kolom=5
kolom = tuple('ABCDE')

data=pd.DataFrame(np.random.randint(1,5, size=(n_baris, n_kolom)),columns=kolom)

data
```

Beginilah cara menggunakan or (|)

```python
data[(data['A']==1)|(data['A']==3)]
```

Or bisa diganti dengan fungsi [.isin()]

```python
data[data['A'].isin([1,3])]
```

Begini cara menggunakan Negasi (Not)[~]

```python
data[~data['A'].isin([1,3])]
```
